```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image #to load image
import numpy as np
```

```python
txt = open("batman.txt","r").read()
txt
```

```python
wc = WordCloud()
wc.generate(txt)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wc,interpolation = "bilinear")
plt.show()
```

```python
wc = WordCloud(background_color="white")
wc.generate(txt)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wc,interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(background_color="white",max_font_size=50,max_words=100)
wc.generate(txt)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wc,interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
custom_mask = np.array(Image.open("cloud.png"))
wc = WordCloud(background_color="white",stopwords=STOPWORDS,mask=custom_mask)
wc.generate(txt)
```

```python
custom_mask = np.array(Image.open("batman.png"))
wc = WordCloud(background_color="white",stopwords=STOPWORDS,mask=custom_mask)
wc.generate(txt)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wc,interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func=image_colors)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wc,interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
custom_mask = np.array(Image.open("batman.png"))
wc = WordCloud(background_color="white",stopwords=STOPWORDS,mask=custom_mask,contour_color="black",contour_width=3)
wc.generate(txt)
```

```python
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func=image_colors)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wc,interpolation = "bilinear")
plt.axis("off")
plt.show()
```
